## Understanding `AutoModelForCausalLM` and Other Model Types

When we import `AutoModelForCausalLM`, we are loading a **Causal Language Model** — the type of model used by GPT-style architectures.  
It uses only the **decoder** part of the Transformer and is designed to **generate text from a prompt**, predicting the next token step by step.

But `AutoModelForCausalLM` is not the only option. There are three main types of models you should know:

1. **`AutoModelForCausalLM`**  
   - **Architecture:** Decoder-only  
   - **Purpose:** Text generation from scratch or from a prompt.  
   - **Examples:** GPT-2, LLaMA, Mistral.

2. **`AutoModelForSeq2SeqLM`**  
   - **Architecture:** Encoder–Decoder  
   - **Purpose:** Transforming an input sequence into an output sequence.  
   - **Examples:** Translation, summarization, text-to-text tasks.  
   - **Examples of models:** T5, BART, mBART.

3. **`AutoModelForMaskedLM`**  
   - **Architecture:** Encoder-only  
   - **Purpose:** Understanding the meaning and relationships between words. Often used for classification, entity extraction, and fill-in-the-blank tasks.  
   - **Examples:** BERT, RoBERTa, DistilBERT.


## Exploring the Tokenizer (Gear)

After downloading the tokenizer and naming it **Gear**,  
let’s see what it looks like and how it works.

```python
text = "I love AI"
tokens = Gear(text)
print(tokens)
print(Gear.convert_ids_to_tokens(tokens["input_ids"]))
```

## Understanding `input_ids` and `attention_mask`

When we give text to the Tokenizer (Gear), it returns two things:

---

### 1. `input_ids`
The tokenizer changes each word (or subword) into a number (an ID from its vocabulary).  
These IDs are just references — the model later converts them into embeddings where similar words have similar vector representations.

Example:  
If the vocabulary assigns:  
- "cat" → `30`  
- "dog" → `45`  
- "door" → `9834`  

Then when we give the sentence:  
"I love AI" → `[40, 1842, 9552]`  
- `40` = "I"  
- `1842` = "love"  
- `9552` = "AI"

---

### 2. `attention_mask`
This tells the model which tokens are real words and which are padding added to make sentences the same length.

Why do we need this?  
Because sentences are not always the same length, and the model (like a stubborn child) prefers all sequences to have equal length.

Example:  
- Sentence 1: 4 real words → `attention_mask`: `[1, 1, 1, 1]`  
- Sentence 2: 3 real words → we add 1 padding token at the end → `attention_mask`: `[1, 1, 1, 0]`

Here:  
- `1` = real token (use it)  
- `0` = padding token (ignore it)


```python
prompt = "Write a detailed emotional short story (about 100 words) about a young man who forgot to eat for 3 days because he was busy building his dream project. Show his physical exhaustion, his thoughts during the process, and end the story with him finally eating and reflecting on his journey."
inputs = Gear(prompt, return_tensors="pt")
```

```python
prompt = "Write a short fictional story about a robot telling its life story."
inputs = Gear(prompt, return_tensors="pt")
```

```python
outputs = V60.generate(
    **inputs,
    max_new_tokens=100,
    temperature=0.7,
    top_p=0.9,
    do_sample=True
)
```

```python
print(Gear.decode(outputs[0], skip_special_tokens=True))
```
